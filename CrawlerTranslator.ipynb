```python
links = ['https://www.wordreference.com/es/en/translation.asp?spen=carne', 'https://www.wordreference.com/es/en/translation.asp?spen=pescado']
print(links)
```

## Solicitudes a la web

```python
import requests

response = requests.get(links[1])

print(dir(response))
```

```python
print(response.text)
```

## Extraccion de informacion del HTML

### Extraccion de la traduccion al ingles

```python

import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')
translation = soup.select('td.ToWrd')[1]
translation.em.extract()

wordEnglish = translation.text
print(wordEnglish)
```

### Extraccion de la traduccion al japones

```python
response = requests.get('https://jisho.org/search/fish')
soup = bs4.BeautifulSoup(response.text, 'html.parser')
wordJapanese = soup.select('div.concept_light-wrapper span.text')[0].text
print(wordJapanese)
```

```python
furigana = soup.select('div.concept_light-wrapper span.furigana span')[0].text
print(furigana)
```
